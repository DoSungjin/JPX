```python
# unnamed: 0 확인 후 삭제
print(df_11.iloc[:,-1].isnull().sum())
print(df_12.iloc[:,-1].isnull().sum())
print(df_13.iloc[:,-1].isnull().sum())
print(df_21.iloc[:,-1].isnull().sum())

df_11.drop(df_11.columns[-1], axis=1, inplace=True)
df_12.drop(df_12.columns[-1], axis=1, inplace=True)
df_13.drop(df_13.columns[-1], axis=1, inplace=True)
df_21.drop(df_21.columns[-1], axis=1, inplace=True)
```

```python
# 각 컬럼 이름 변경
df_11 = df_11.rename(columns={'일자': 'date',
                    '시가': 'open',
                    '고가': 'high',
                    '저가': 'low',
                    '종가': 'close',
                    'TRIX 기준선 매수': 'TRIX_buy',
                    '매수청산': 'TRIX_buy_clear',
                    '매도': 'TRIX_sell',
                    '매도청산': 'TRIX_sell_clear'})

df_12 = df_12.rename(columns={'일자': 'date',
                    '시가': 'open',
                    '고가': 'high',
                    '저가': 'low',
                    '종가': 'close',
                    'MACD 기준선 매수': 'MACD_buy',
                    '매수청산': 'MACD_buy_clear',
                    '매도': 'MACD_sell',
                    '매도청산': 'MACD_sell_clear'})

df_13 = df_13.rename(columns={'일자': 'date',
                    '시가': 'open',
                    '고가': 'high',
                    '저가': 'low',
                    '종가': 'close',
                    'TRIX 크로스 매수': 'TRIX_C_buy',
                    '매수청산': 'TRIX_C_buy_clear',
                    '매도': 'TRIX_C_sell',
                    '매도청산': 'TRIX_C_sell_clear'})

df_21 = df_21.rename(columns={'일자': 'date',
                    '시가': 'open',
                    '고가': 'high',
                    '저가': 'low',
                    '종가': 'close',
                    'TRIX 기준선 매수': 'TRIX_buy',
                    '매수청산': 'TRIX_buy_clear',
                    '매도': 'TRIX_sell',
                    '매도청산': 'TRIX_sell_clear'})

# 각 데이터프레임의 date 컬럼을 datetime으로 변경
df_11['date'] = pd.to_datetime(df_11['date'])
df_12['date'] = pd.to_datetime(df_12['date'])
df_13['date'] = pd.to_datetime(df_13['date'])
df_21['date'] = pd.to_datetime(df_21['date'])
```

```python
# df_11, df_12, df_13를 하나의 데이터프레임으로 합치기
df_1 = pd.merge(df_11, df_12, how='inner', on=['date', 'open', 'high', 'low', 'close'])
df_1 = pd.merge(df_1, df_13, how='inner', on=['date', 'open', 'high', 'low', 'close'])
df_1
```

```python
#df_1에서 date 컬럼을 index로 변경
df_1 = df_1.set_index('date')
df_1
```

```python
# df_1에서 TRIX_buy_clear와 TRIX_sell가 같은지 확인
print(f'len of df_1 {len(df_1.TRIX_buy_clear==df_1.TRIX_sell)}')
print(f'TRIX_buy_clear와 TRIX_sell 같은지 확인 {list(df_1.TRIX_buy_clear==df_1.TRIX_sell).count(True)}')

for i in range(len(df_1)):
    if df_1.TRIX_buy_clear[i] != df_1.TRIX_sell[i]:
        print(df_1.TRIX_buy_clear[i])
        print(df_1.TRIX_sell[i])
```
