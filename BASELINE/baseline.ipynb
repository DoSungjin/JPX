```python
df_all_0 = df_all.fillna(0)
# int로 변경
df_all_0['TRIX_buy'] = df_all_0['TRIX_buy'].astype(float)
df_all_0['TRIX_buy_clear'] = df_all_0['TRIX_buy_clear'].astype(float)
```

```python
# Backtest
# 1. TRIX 기준선 매수
# 2. MACD 기준선 매수
# 3. TRIX 크로스 매수
# 4. TRIX 기준선 매수 + MACD 기준선 매수
# 5. TRIX 크로스 매수 + MACD 기준선 매수
# 6. TRIX 기준선 매수 + MACD 기준선 매수 + TRIX 크로스 매수

# 1. TRIX 기준선 매수
# 1-1. TRIX_buy가 None이 아닌 경우
mk_TRIX_buy_list = []
for i in range(len(df_all_0)):
    if df_all_0.TRIX_buy[i] == 0.0 :
        mk_TRIX_buy_list.append(False)
    else:
        mk_TRIX_buy_list.append(True)

df_all_0['TRIX_buy_mk_true'] = mk_TRIX_buy_list

# 1-2, TRIX_buy_clear가 None이 아닌 경우
mk_TRIX_buy_clear_list = []
for i in range(len(df_all_0)):
    if df_all_0.TRIX_buy_clear[i] == 0.00000 :
        mk_TRIX_buy_clear_list.append(False)
    else :
        mk_TRIX_buy_clear_list.append(True)

df_all_0['TRIX_buy_clear_mk_true'] = mk_TRIX_buy_clear_list
```
